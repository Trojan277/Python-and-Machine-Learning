- Veri setinde yer alan birden fazla veri aslinda tek bir veri ile temsil edilebillir
    - Boy ve kilo : Vucut kitle endeksi
    - Enlem ve boylam degisimi : Alinan mesafe
- Icsel boyut (intrinsic dimension) : Veri setinde bulunmasi gereken ozellik (feature) sayisi
    - Boyut azaltmanin temelindeki mantik
    - Bir veri setinin en az sayida veri ile temsil edilmesi



```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
area = data[:,0]
lenght = data[:,3]
width = data [:,4]
geometry = data[:,3:5]
```

```python
import matplotlib.pyplot as plt
plt.scatter(width,lenght)
plt.xlabel('Width of Kernel')
plt.ylabel('Lenght of Kernel')
plt.axis('equal') # plt.axis ('scaled') : olceklendirir
plt.show()
from scipy.stats import pearsonr
correlation, pvalue = pearsonr(lenght,width)
print(correlation)
from sklearn.decomposition import PCA
model = PCA()
pca_features = model.fit_transform(geometry)
x = pca_features[:,0]
y = pca_features[:,1]
plt.axis('equal')
plt.scatter(x,y)
plt.show()
correlation, pvalue = pearsonr(x,y)
print(correlation)
```

```python
# Varyansin yonelimini belirlemek
print(model.components_)
```

```python
model = PCA()
model.fit(geometry)
mean = model.mean_
start = model.components_[0,:]

plt.scatter(geometry[:,0],geometry[:,1])
plt.xlabel('Width of Kernel')
plt.ylabel('Lenght of Kernel')
plt.axis('equal')

plt.arrow(mean[0], mean[1], start[0], start[1], color = 'red', width = 0.05)

plt.show()
```

```python
# 3 boyutlu dagilim grafigi

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
x = fig.add_subplot(projection = '3d')

x.scatter(area, lenght, width)
x.set_xlabel('Alan')
x.set_ylabel('Uzunluk')
x.set_zlabel('Genislik')

plt.show()
```

```python
# Varyanslarin dagilimini/etkisini belirlemek

data2 = data[:,0:7]
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

scaler = StandardScaler()
pca = PCA()

pipeline = make_pipeline(scaler, pca)
pipeline.fit(geometry)

features = range(pca.n_components_)

plt.bar(features, pca.explained_variance_)
plt.xlabel('Feature')
plt.ylabel('Variance')
plt.xticks(features)
plt.show()
```
