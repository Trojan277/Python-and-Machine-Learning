Veriler kucukten buyuge siralandiktan sonra, seti 4 esit parcaya bolen degerlere **yuzdelikler** denir. Yani veriyi %25'lik dilimlere ayirir.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

veri = pd.read_csv('2008_all_states.csv')
sns.set()
x = np.sort(veri['dem_share'])
y = np.arange(1, len(x) + 1) / len(x)
plt.plot(x,y, marker = '.', linestyle='none')
plt.xlabel('Oy orani')
plt.ylabel('ECDF')
plt.title('2008 Secimleri Oy Analizi')
plt.margins(0.05)
plt.show().
```

```python
# Yuzdelik Hesaplama
yuzdelikler = np.percentile(veri['dem_share'], [25,50,75])
print(yuzdelikler)
```

```python
# Box plot olusturmak
sns.boxplot(x = 'east_west', y= 'dem_share', data=veri)
plt.xlabel('Eyalet (Bati - Dogu)')
plt.ylabel('Alinan Oy Oranlari')
plt.title('20008 Secim Analizleri')
plt.rcParams['figure.figsize'] = [30,15]
plt.show()
```

```python
sns.boxplot(x = 'state', y= 'dem_share', data=veri)
plt.xlabel('Eyalet (Bati - Dogu)')
plt.ylabel('Alinan Oy Oranlari')
plt.title('20008 Secim Analizleri')
sns.swarmplot(x = 'state', y= 'dem_share', data=veri)
plt.show()
```
