- fit_transform() metodu kullanilir
    - fit() metodu yoktur
    - transform() metodu yoktur
    - Yeni verilere eklenemez

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
wheat = data[:,:-1]
location = data[:,-1]
```

```python
from sklearn.manifold import TSNE

model = TSNE(learning_rate = 200)
tsne = model.fit_transform(wheat)

x = tsne[:,0]
y = tsne[:,1]

import matplotlib.pyplot as plt

plt.scatter(x,y, c = location)
plt.show()
```
