```python
def mod3(sayi1,sayi2,sayi3): # Function Header
    """Verilen sayilarin ucten kalanini hesaplar"""
    
    yeni_sayi1 = sayi1 % 3
    yeni_sayi2 = sayi2 % 3
    yeni_sayi3 = sayi3 % 3
    
    return (yeni_sayi1,yeni_sayi2,yeni_sayi3)
```

```python
mod3(14,57,10)
```

```python
def mod4(x1,x2,x3): # Function Header
    """Verilen sayilarin dortten kalanini hesaplar""" # Fonksiyonun amacini belirler

    
    def icmod4(x):
        return x % 4
    return (icmod4(x1),icmod4(x2),icmod4(x3))
```

```python
mod4(5,7,12)
```

```python
# Fonksiyon dondurme

def usalma(n):
    "Sayinin ussunu alir"
    
    def icusalma(x):
        sayi = x ** n
        return sayi
    return icusalma
```

```python
kare = usalma(2)
```

```python
kare(24)
```

```python
# nonlocal
def dis():
    
    x = 10
    
    def ic():
        nonlocal x 
        x = 30
        print(x)
    ic()
    print(x)
```

```python
dis()
```

Soru 1 : **dış( ) ve iç( ) fonksiyonlarının çıktısı nedir?**

```python
def dış():
    x = "ali"
 
    def iç():
        nonlocal x
        x = "ayşe"
        print("iç:", x)
 
    iç()
    print("dış:", x)
 
dış()
```
