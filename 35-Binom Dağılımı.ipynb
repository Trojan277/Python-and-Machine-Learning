# Binom Dagilimi

Olasilik kurami ve istatistik bilim kollarinda, binom dagilimi n sayida iki karegori (yani basari/basarisizlik, evet/hayir, 1/0 vb) sonucu veren hesaplamalardir

# Olasilik Kutle Fonksiyonu (Probability Mass Function | PMF)

Binom dagilimi icin en basit ornek bir zarin 10 defa atilip kac tane 6 elde edildiginin sayilmasidir. Bu rassal sayinin (yani 10 deneyde kac tane 6 elde edilmesi) dagilimi, n=10 ve p=1/6 parametreleri olan bir binom dagilimidir. (Tekduze | Discrete Uniform PMF)
