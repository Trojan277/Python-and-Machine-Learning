```python
import numpy as np
np.random.seed(123)
tura = [0]
for x in range(10):
    para = np.random.randint(0,2)
    tura.append(tura[x] + para)
print(tura)
```

```python
# 100 kere 10 defa para atma islemi yapma
tura_liste = []
for x in range(100000):
    tura = [0]
    for x in range(10):
        para = np.random.randint(0,2)
        tura.append(tura[x] + para)
    tura_liste.append(tura[-1])
print(tura_liste)
```

```python
import matplotlib.pyplot as plt
plt.hist (tura_liste, bins = 10)
plt.show()
```

```python
import numpy as np
np.random.seed(123)

tum_zar_liste = []
for x in range(10):
    zar_liste = [0]
    for x in range(100):
        seviye = zar_liste[-1]
        zar = np.random.randint(1,7)
        if zar <= 2:
            seviye = max(0, seviye -1)
        elif zar < 6:
            seviye = seviye + 1
        else:
            seviye = seviye + np.random.randint(1,7)
            
        if np.random.rand() <= 0.001:
            seviye = 0 
        zar_liste.append(seviye)
    tum_zar_liste.append(zar_liste) 
print(tum_zar_liste)
```

```python
import matplotlib.pyplot as plt
dizi = np.array(tum_zar_liste)
dizi2 = np.transpose(dizi)
plt.plot(dizi2)
plt.show()
```
