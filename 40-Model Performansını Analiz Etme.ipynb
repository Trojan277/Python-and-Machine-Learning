Hassasiyet (Accuracy): Dogru tahmin edilen verilerin orani ( Dogru veri / Toplam veri )

## Modelin hassasiyeti nasil belirlenir?

- fit( ) kullanilabilir
- Egitim verisi **KESINLIKLE KULLANILMAMALIDIR!** 
- Test veri seti kullanilmalidir.

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
iris = datasets.load_iris()
x = iris.data
y = iris.target
veri = pd.DataFrame(x, columns = iris.feature_names)
```

```python
veri.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state= 100, stratify = y)
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train, y_train)
accuracy = knn.predict(x_test)
print(accuracy)
knn.score(x_test, y_test)
```

1.0

```python
# Performansi Grafik ile Analiz Etmek
neighbors = np.arange(1,10)
print(neighbors)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Enumerate
# sayilar = (12,54,87,96,12)
# for x,y in enumerate(sayilar):
#    print(x,y)         # Degerleri index sayilari ile beraber verir

for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test)
    
plt.plot(neighbors, train_accuracy, label = 'Train Data Accuracy')
plt.plot(neighbors, test_accuracy, label = 'Test Data Accuracy')
plt.title('Iris Flower MLM')
plt.legend()
plt.xlabel('# Neighbors')
plt.ylabel('Accuracy')
plt.show
```
