- Hiyerarsik kumeleme (SciPy)
- t-SNE: 2 boyutlu veri seti haritasi / grafigi

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
wheat = data[:,:-1]
location = data[:,-1]
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
import matplotlib.pyplot as plt
```

```python
merge = linkage(wheat, method = 'complete')

plt.figure(figsize = (20,20))
dendrogram(merge, labels = location, leaf_rotation = 90, leaf_font_size = 12, orientation = 'left')
plt.show()
```
