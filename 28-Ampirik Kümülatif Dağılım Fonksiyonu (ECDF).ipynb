```python
x = np.sort(veri['rep_votes'])
y = np.arange(1, len(x) +1) / len(x) 
plt.plot(x,y, marker = '.', linestyle = 'none') # marker = '.' komutu grafigi nokta nokta yapar, linestyle = 'none' komutu verileri birbirine baglayan cizgileri kaldirir
plt.xlabel('Oy Orani')
plt.ylabel('ECDF')
plt.title('2008 Secimleri Oy Analizi')
plt.margins(0.1)

x1 = np.sort(veri['dem_votes'])
y1 = np.arange(1, len(x) +1) / len(x) 
plt.plot(x1,y1, marker = '.', linestyle = 'none') # marker = '.' komutu grafigi nokta nokta yapar, linestyle = 'none' komutu verileri birbirine baglayan cizgileri kaldirir
plt.xlabel('Oy Orani')
plt.ylabel('ECDF')
plt.title('2008 Secimleri Oy Analizi')
plt.margins(0.1)
plt.legend(('Cumhuriyetci', 'Demokrat'))
plt.show()
```

```python
def secim(veri):
    "ECDF grafigi icin x ve y degerlerini hesaplar"
    n = len(veri) # toplam veri sayisini hesaplama
    x = np.sort(veri) # x eksenini tanimlama
    y = np.arange(1 , n + 1) / n
    
    return x,y
```

```python
x_dem, y_dem = secim(veri['dem_votes'])
x_rep, y_rep = secim(veri['rep_votes'])

plt.plot(x_dem, y_dem)
plt.plot(x_rep, y_rep)
plt.show()
```
