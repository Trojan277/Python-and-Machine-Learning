```python
# DataFrame'e cevirme (Veri cercevesi)
x = iris.data
y = iris.target
veri = pd.DataFrame(x, columns = iris.feature_names)
print(veri.head())
```

```python
# Grafisel Kessifsel Veri Analizi
pd.plotting.scatter_matrix(veri, c = y, figsize = [10,10], s = 100, marker = 'x')
plt.show()
```
