```python
# if, elif, else

liste = [sayi ** 2 for sayi in range(1,51)]
print(liste)
```

```python
liste2 = [sayi ** 2 for sayi in range(1,51) if sayi % 2 == 0]
print(liste2)
```

```python
liste3 = [sayi ** 2 if sayi % 2 == 0 else str(sayi) +'Tam Bolunmuyor' for sayi in range (1,51)]
print(liste3)
```

```python
# Dict
# [] yerine {}

liste4 = {sayi: sayi **2 for sayi in range(1,11)}
print(liste4)
```
