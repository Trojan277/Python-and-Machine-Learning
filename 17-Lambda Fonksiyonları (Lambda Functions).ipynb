# Lambda Fonksiyonlari
Lambda fonksiyonlari herhangi bir isim tanimlamasi yapilmadan kullanilan anonim fonksiyonlardir. Normal sartlarda bir fonksiyon yazmak icin def kullanilirken anonim fonksiyonlar icin lambda kullanilir.

## Lambda Fonksiyonlarinin kullanimi

- lambda arguments(s): expression
- lambda parametre(ler): yapilacak islem

## Lambda Fonksiyonlari ne zaman kullanilir?

Kisa sureli ve basit islemler icin yazimi daha pratik olmasi nedeniyle lambda fonksiyonlari tercih edilebilir. Bununla birlikte lambda fonksiyonlari ön tanımlı (build-in) olarak bulunan filter(), map() gibi fonksiyonlarla birlikte de kullanilirlar.

```python
iki_kat = lambda x: x*3
iki_kat(5)
```

```python
us_alma = lambda x, y: x ** y
us_alma(4,5)
```

```python
liste = [4,7,11,14,18]

ucten_kalan = list(filter(lambda sayi: (sayi % 3 == 0), liste))
```

```python
uc_ussu = map(lambda sayi: sayi ** 3, liste)
```

```python
uc_ussu
```

```python
list(uc_ussu)
```
