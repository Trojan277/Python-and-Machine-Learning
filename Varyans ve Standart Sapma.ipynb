# Varyans ve Standrt Sapma

**Varyans**, bir olasilik dagilimini veya orneklem icin istatistiksel yayilimin, mumkun butun degelerin beklenen deger veya **ortalamadan uzakliklarinin karelerinin ortalamasi** seklinde bulunan bir olcuttur.

**Standart Sapma** bir olasilk dagilimi veya orneklem icin istatiksel yayilimin, **veri degelerinin yayiliminin ozetlenmesi** icin kullanilan bir olcuttur.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_csv('2008_swing_states.csv')
sns.set()
sns.swarmplot(x = 'state', y = 'dem_share', data = data)
plt.xlabel('Eyalet')
plt.ylabel('Demokrat Oy Orani')
plt.title('Demokratlarin Swing States Oylari (2008 Secimleri)')
plt.show()
```

```python
import numpy as np
data2 = data[data['state'] == 'FL'] # parantezin disindaki data yazilmazsa true - false gosterilir
data3 = data2['dem_share']
```
