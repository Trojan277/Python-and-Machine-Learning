- Boyut azaltma (dimension reduction)
- Veri setindeki "onemsiz" verileri elemine etmek
    - Bu tur veriler siniflandirma, regresyon performansini azaltir
- PCA = Principal Component Analysis
- En temel ve yaygin boyut azalma yontemidir
    - Veriler "de-korele" edilir
        - Veriler uzerinde herhangi bir degisiklik yapilmaz
        - Verileri neksenler dogrultusunda siralanir (mean = 0)
        - Veri/bilgi kaybi meydana gelmez
    - Boyut azaltma

```python
import numpy as np
data = np.loadtxt('seeds_dataset.txt')
lenght = data[:,3]
width = data [:,4]
geometry = data[:,3:5]
```

```python
import matplotlib.pyplot as plt

plt.scatter(width,lenght)
plt.xlabel('Width of Kernel')
plt.ylabel('Lenght of Kernel')
plt.axis('equal') # plt.axis ('scaled') : olceklendirir
plt.show()
from scipy.stats import pearsonr
correlation, pvalue = pearsonr(lenght,width)
print(correlation)
```


0.8604149377143469

```python
print(geometry)
```

```python
from sklearn.decomposition import PCA

model = PCA()

pca_features = model.fit_transform(geometry)

x = pca_features[:,0]
y = pca_features[:,1]
plt.axis('equal')
plt.scatter(x,y)
plt.show()
correlation, pvalue = pearsonr(x,y)
print(correlation)
```


6.938893903907228e-18
