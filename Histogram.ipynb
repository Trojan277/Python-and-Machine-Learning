```python
# maplotlib'in indirilmesi
# pandas'in indirilmesi

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('2008_swing_states.csv')

plt.hist(data['dem_share']) # Histogram olusturma
plt.xlabel('Demokrat Oy Orani') # x ekseni basligi
plt.ylabel('Ilce') # y ekseni basligi
plt.title('Demokratlarin Swing States Oylari (2008 Secimleri)')
plt.show()
```

```python
plt.hist(data['dem_share'], bins = 15) # Histogram olusturma
plt.xlabel('Demokrat Oy Orani') # x ekseni basligi
plt.ylabel('Ilce') # y ekseni basligi
plt.title('Demokratlarin Swing States Oylari (2008 Secimleri)')
plt.show()
```
