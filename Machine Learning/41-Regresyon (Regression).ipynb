Hedef degisken (target variable) bir sabit degil aksine degiskendir. 
- Bir evin fiyati 
- Bir ulkenin toplam gayri safi milli hasilasi
- Bir isletmenin aylik geliri

```python
import pandas as pd
data = pd.read_csv('boston.csv')
```

```python
data.head()
```

```python
# scikit-learn
# RMve MEDV degerleri ayri ayri saklanmalidir
X = data.drop('medv', axis = 1).values # target veriable haric tum veriler
y = data['medv'].values # sadece target veriable
```

```python
y
```

```python
X
```

```python
# Tek bir parametre kullanmak
room_number = X[:,5]
print(room_number)
```

```python
room_number = room_number.reshape(-1,1)
print(room_number)
```

```python
room_number = X[:,5]
room_number = room_number.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
# Grafik olustur
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.scatter(room_number,y)
plt.title('Oda Sayisis ile Ev Fiyati Arasindaki Iliski')
plt.xlabel('Oda Sayisi')
plt.ylabel('Ev Fiyati(x $1000)')
plt.show()
```

```python
# Regresyon Modeli Olusturma
import numpy as np
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(room_number, y)

regression_line = np.linspace(min(room_number),max(room_number))

plt.scatter(room_number, y)
plt.title('Oda Sayisis ile Ev Fiyati Arasindaki Iliski')
plt.xlabel('Oda Sayisi')
plt.ylabel('Ev Fiyati(x $1000)')
plt.plot(regression_line,reg.predict(regression_line), color = 'red', linewidth = 3)
plt.show()
```
