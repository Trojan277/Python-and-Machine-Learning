**Pearson Korelasyon Katsayisi**

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
sns.set()
veriler = pd.read_csv('2008_all_states.csv')
```

```python
sns.scatterplot(x='total_votes', y='dem_share', data = veriler)
plt.xlabel('Toplam Oy Sayisi')
plt.ylabel('Demokrat Oy Sayisi')
plt.title('2008 Secim Analizi')
plt.show()
```

```python
import numpy as np
kovaryans = np.cov(veriler['total_votes'], veriler['dem_share'])
print(kovaryans)
```

```python
# Pearson Korelasyon Katsayisi
def pearson(x,y):
    """Pearson korelasyon katsayisini hesaplar"""
    korelasyon = np.corrcoef(x,y)
    return korelasyon[0,1]
```

```python
p = pearson (veriler['total_votes'], veriler['dem_share'])
print(p)
```
