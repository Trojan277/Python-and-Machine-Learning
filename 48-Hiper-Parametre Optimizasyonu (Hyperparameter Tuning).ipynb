## Hiper-Parametre Secimi 
- Modelin basarisi hiper-parametre secimine baglidir. 
- Farkli degerleri uydurmak ve en verimlisini secmek izlenecek yoldur. 
- Asiri uymanin (overfitting) onune gecmek icin capraz dogrulama (cross validation) yapilmalidir 

Bu islem Hiper-Parametre Optimizasyonu (Hyperparameter Tuning) olarak adlandirilir

## Grid Aramasi ve Capraz Dogrulama (Grid Search and Cross Validation) 
- Farkli parametrelerin, farkli degerlerde performansi karsilastirilir ve en yuksek verimin alindigi degerler belirlenir.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
data = pd.read_csv('diabetes.csv')
X = data.drop(['Outcome'], axis = 1)
y = data ['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 55)
```

```python
from sklearn.model_selection import GridSearchCV
space = np.logspace(0,2, 50)
parameters = {'C': space}
logreg =  LogisticRegression(max_iter = 1000)

logreg.GSCV = GridSearchCV(logreg,parameters,cv = 5)
logreg.GSCV.fit(X_train, y_train)
print('Lojistik Regresyon Parametresi: {}'.format(logreg.GSCV.best_params_))
print('En iyi performans degeri: {}'.format(logreg.GSCV.best_score_))
```
