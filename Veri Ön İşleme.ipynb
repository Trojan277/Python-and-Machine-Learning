```python
import pandas as pd
data = pd.read_csv('gm_2008_region.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data.boxplot('life','Region', rot = 45)
plt.title('Bolgelere Gore Ortalama Yasama Suresi')
plt.xlabel('Bolge')
plt.ylabel('Yil')
plt.suptitle('')
plt.show()
```

```python
# Dummy Variable Olusturma
data_dummy = pd.get_dummies(data)
print(data_dummy.columns)
data_dummy.head()
```

```python
data_dummy2 = pd.get_dummies(data, drop_first = True)
print(data_dummy2.columns)
data_dummy2.head()
```

```python
# Regresyon 

from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X = data_dummy2.drop(['life'], axis = 1)
y = data_dummy2['life']

X_train, y_train, X_test, y_test = train_test_split(X, y, test_size = 0.20, random_state = 30)

ridge = Ridge(alpha = 0.5, normalize = True)
ridge_cv = cross_val_score(ridge, X, y, cv = 5)
print(ridge_cv)
```
