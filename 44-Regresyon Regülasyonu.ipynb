## Ridge Regresyonu

- Hata fonksiyonu = En kucuk kareler yontemi (Ordinary Least Squares Method, OLS) + alfa
- Alfa : Karar verilmesi gerken bir parametre ( Komsu sayisi gibi )
    - 0 alinirsa normal OLS fonksiyonu (Overfitting)   
    - Cok yuksek bir deger alinirsa (Underfitting) 
    - Model karmasikligini belirler

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import pandas as pd
data = pd.read_csv ('boston.csv')
# scikit - learn
# RM ve MEDV degerleri ayri ayri saklanmalidir
X = data.drop('medv', axis =1).values # target variable haric tum veriler
y = data['medv'].values
data.head()
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 30)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha = 0.1, normalize = True)
ridge.fit(X_train, y_train)
ridge_prediction = ridge.predict(X_test)
ridge.score(X_test, y_test)
```
